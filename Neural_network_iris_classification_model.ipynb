```python
# -*- coding: utf-8 -*-
# !/usr/bin/python3
# author by : yuxiangShi
# tell  by: 18538187569
# desc by: 通过构建model的方式搭建基于神经网络的鸢尾花分类

# 1.导入基础类包
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras import Model
import numpy as np
from sklearn import datasets
# 2.导入包资源
datasets = datasets.load_iris()
# 3.构建基础特征和标签
x_train = datasets.data # 特征
y_train = datasets.target # 标签
# 4.创建model类型
class IrisModel(Model):
  def __init__(self):
    super(IrisModel,self).__init__()
    #全连接层网络结构（激活函数为softmax；正则化为l2）
    self.d1 = Dense(3, activation='softmax', kernel_regularizer=tf.keras.regularizers.l2())

  def call(self,x):
    y = self.d1(x)
    return y

model = IrisModel()

# 5.构建神经网络
# 配置神经网络的训练方法，优化器为sgd；loss损失函数为；metrics评测指标为：y_是数值，y是独热码
model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=0.1),
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),
              metrics=['sparse_categorical_accuracy'])

# 6.训练模型
model.fit(x_train,y_train,batch_size=32,epochs=500,validation_split=0.2,validation_freq=20)
# 7.输出统计和数据
model.summary()
```
